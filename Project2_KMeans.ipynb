```python
# Classification by KMeans Clustering
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('/content/drive/MyDrive/Python_AIML/fruit_data_with_colors.txt',sep='\t')
df
```

```python
X=df.iloc[:,[3,6]].values
y=df.iloc[:,0].values
```

```python
from sklearn.cluster import KMeans
model=KMeans(n_clusters=4,random_state=13)
model.fit(X)
```

```python
y_p=model.predict(X)
# for i in range(len(y_p)):
#   y_p[i]+=1
y_p
```

```python
print(np.unique(y,return_counts=True))
np.unique(y_p,return_counts=True)
```

```python
plt.figure(figsize=(12,5))
plt.scatter(X[y_p==0,0],X[y_p==0,1],label="Cluster 0")
plt.scatter(X[y_p==1,0],X[y_p==1,1],label="Cluster 1")
plt.scatter(X[y_p==2,0],X[y_p==2,1],label="Cluster 2")
plt.scatter(X[y_p==3,0],X[y_p==3,1],label="Cluster 3")
plt.scatter(model.cluster_centers_[:,0],model.cluster_centers_[:,1],label="Centroids",color="cyan",s=200,alpha=0.5)
plt.xlabel("Mass")
plt.ylabel("Color_Score")
plt.title("Fruits Analysis")
plt.legend()
plt.show()
```
