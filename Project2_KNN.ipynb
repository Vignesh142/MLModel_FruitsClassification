```python
# Classification by KNN
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('/content/drive/MyDrive/Python_AIML/fruit_data_with_colors.txt',sep='\t')
df
```

```python
from sklearn.neighbors import KNeighborsClassifier
model=KNeighborsClassifier(n_neighbors=5,metric='euclidean')
```

```python
# there is more accuracy in taking parameters as width and color_score
X=df.iloc[:,[4,6]].values
y=df.iloc[:,1].values
```

```python
model.fit(X,y)
```

```python
y_p=model.predict(X)
print(y)
y_p
```

```python
a=np.unique(y_p)
a
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix, classification_report
accuracy_score(y_p,y)
```

```python
confusion_matrix(y_p,y)
```

```python
print(classification_report(y_p,y))
```

```python
x_test=[[7.5,0.70]]
```

```python
y_predict=model.predict(x_test)
y_predict
```

```python
plt.figure(figsize=(14,6))
plt.scatter(X[y==a[0],0],X[y==a[0],1],label=a[0],marker="o",color="red")
plt.scatter(X[y==a[1],0],X[y==a[1],1],label=a[1],marker="o",color="k")
plt.scatter(X[y==a[2],0],X[y==a[2],1],label=a[2],marker="o",color="green")
plt.scatter(X[y==a[3],0],X[y==a[3],1],label=a[3],marker="o",color="blue")
plt.scatter(x_test[0][0],x_test[0][1],label='Predicted',marker="*",color="brown",s=500)
plt.legend(loc="upper left")
plt.xlabel("Weight")
plt.ylabel("color_score")
plt.title("Fruit Analysis")
plt.grid()
plt.show()
```
